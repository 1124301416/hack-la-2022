```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
import pickle
nltk.download('punkt')
from wordcloud import WordCloud
```

```python
discussion = pd.read_csv('discussion_posts.csv')
```

```python
with open('stopwords-Copy1.pkl', 'rb') as f:
    stopwords = pickle.load(f)
```

```python
print(type(stopwords), len(stopwords), stopwords, sep="\n")
```

```python
with open('Discussion.txt', encoding='utf-8') as f:
    words = nltk.word_tokenize(f.read().lower())
```

```python
print(words)
```

```python
words2 = [] # our accumulator list

for w in words:
    if w not in stopwords and len(w) > 1:
        words2.append(w)

print(len(words))
print(len(words2))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wordcloud = WordCloud(width=800, height=400).generate(" ".join(words2))
plt.figure(figsize=(20,10)) # set up figure size
plt.imshow(wordcloud) # word cloud image show
plt.axis("on") # turn on axis
plt.savefig('my_word_cloud.png') # save as PNG file
plt.savefig('my_word_cloud.pdf') # save as PDF file
plt.show()  # show in Jupyter notebook
```
